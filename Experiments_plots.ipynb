```python
from torch.utils.tensorboard import SummaryWriter
import matplotlib.pyplot as plt
from tensorboard.backend.event_processing.event_accumulator import EventAccumulator
import numpy as np

writer = SummaryWriter()
```

# Experiment 1

```python
event_acc = EventAccumulator('/home/gm3044/Practical_Deep_Learning_Project/NormalBase')
event_acc.Reload()
# Show all tags in the log file
print(event_acc.Tags())
```

```python
_, _, Loss = zip(*event_acc.Scalars('Loss/Train'))
_, _, val_Loss = zip(*event_acc.Scalars('Loss/Eval'))
_, Epochs, _ = zip(*event_acc.Scalars('LR'))

Train_loss = list(Loss)
Val_loss = list(val_Loss)
Steps = list(Epochs)
```

```python
Loss = []
Val_Loss = []
for i in range(len(Train_loss)-100):
    Loss.append(Train_loss[i+100])
for i in range(len(Val_loss)-100):
    Val_Loss.append(Val_loss[i+100])    
```

```python
Loss = []
for i in range(0, len(Train_loss),100):
    Loss.append(Train_loss[i])
Val_Loss = []
for i in range(0, len(Val_loss),79):
    Val_Loss.append(Val_loss[i])
```

```python
epochs = range(0, 100)

plt.figure(figsize=(10,5))
plt.title("Experiment 1: Training and Validation Loss")
plt.plot(epochs, Val_Loss,label="Validation Loss")
plt.plot(epochs, Loss,label="Training Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend(loc='best')
plt.show()
```
